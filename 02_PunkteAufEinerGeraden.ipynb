# Punkte auf einer Geraden

Auf einer Geraden können sich beliebig viele Punkte befinden. Durch Rechnung kann bestimmt werden, ob ein Punkt auf einer Geraden liegt, beziehungsweise ob eine Gerade durch einen bestimmten Punkt im Koordinatensystem geht. 
Dafür müssen die Koordinaten des Punktes P in die Geradengleichung eingesetzt werden. Steht nach dem Ausrechnen der Terme auf der Linken und Rechten Seite die selbe Zahl, liegt der Punkt auf der Geraden.

### Rechenbeispiel

**Gegeben**: Gerade g: $y = 1,5x + 2,5$ ; $A(1|4)$; $B(-1,5|1,5)$

**Gesucht**: Liegen $A$ und $B$ auf der Geraden $g$?

**Lösung**: 
$A$: 

$x_A = 1$ und $y_A = 4$

=> $A$ in $g$: 

$4 = 1,5 * 1 + 2,5$

$4 = 4$

-> **wahr** $A \in g$





$B$: 

$x_B = -1,5$ und $y_B = 1,5$

=> $B$ in $g$: 

$-1,5 = 1,5 * (-1,5) + 2,5$

$-1,5 = (-2,25) + 2,5$

$-1,5 = 0,25$

-> **falsch** $B \notin g$

## Codebeispiel

Die im Rechenbeispiel vorgestellte Rechnung wird im folgenden Code angezeigt. 

Die Variable **m_1** entspricht dabei Steigung "m", **t_1** entspricht dem y-Achsenabschnitt t.

Der Punkt A ist durch die Variablen **x_A** und **y_A**, Punkt B durch die Variablen **x_B** und **y_B**. 


## Aufgabe

Überprüfe durch Ändern der Variablen im Code, ob der Punkt $A (2|3)$ und der Punkt $B (-1|0)$ auf der Geraden $y = 2x + 2$ liegen. (Einer der Punkte sollte auf der Geraden liegen!)

Wenn du für eine dieser Variablen andere Werte angibst und auf "Run" drückst (oben in der Mitte), verändern sich die Punkte sowie die Gerade. 

<span style='color:red'> **Beachte! Wenn du die Werte veränderst, können sich die x- und y-Achse verschieben!** </span>

```python
import matplotlib.pyplot as plt

xs = []
ys_1 = []

# Steigung der Geraden g
m_1 = 1.5
# y-Achsenabschnitt der Geraden g
t_1 = 2.5

# Punkt A
x_A = 1
y_A = 4

# Punkt B
x_B = -1.5
y_B = 1.5

for x in range(-3, 3):
    xs.append(x)
    ys_1.append((m_1 * x) + t_1)
    
plt.plot(xs, ys_1, "r-") # Rote Gerade
plt.plot(x_A, y_A, marker='o', markersize='10', color="blue")
plt.text(x_A - 0.2, y_A + 0.3, "A")
plt.plot(x_B, y_B, marker='o', markersize='10', color="blue")
plt.text(x_B - 0.25, y_B + 0.15, "B")
plt.grid(True)
plt.show()
```
