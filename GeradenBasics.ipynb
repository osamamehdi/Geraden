# Geradengleichungen

Allgemeine Formel $y = m * x + t$

Dabei steht m für die **Steigung** der Geraden.

Der **y-Achsenabschnitt t**, gibt an welchen Wert die Gerade für **$ x = 0 $** hat.

## Beispiel

Folgendes Beispiel beschreibt eine einfache Gerade. 

Die Variablen 

$m$, $x$ und $t$ 

sind Teil der Geradengleichung.

Für $t = 0$ entsteht eine Gleichung mit der Form $y = m*x$. Geraden mit dieser Form verlaufen durch den 
Ursprung $ O (0|0)$ und werden **Ursprungsgeraden** genannt.

```python
import matplotlib.pyplot as plt

xs = []
ys_1 = []

m_1 = 1
t_1 = 0

for x in range(0, 3):
    xs.append(x)
    ys_1.append((m_1 * x) + t_1)
    
plt.plot(xs, ys_1, "r-") # Rote Gerade
plt.grid(True)
plt.show()
```

## Steigung

Die Variable **"m"** beschreibt die Steigung der Gerade. Diese lässt sich mit der Formel 

$m = \frac{y_2 - y_1}{x_2 - x_1}$ 

berechenen.

## Beispiel

Folgender Code beschreibt zwei Geraden. 

**m_1** und **m_2** stehen für die Steigungen der Geraden.

Anhand der unten eingeblendeten Geraden erkennst du, wie sich die Steigung auf die Geraden auswirkt.

```python
import matplotlib.pyplot as plt

xs = []
ys_1 = []
ys_2 = []

m_1 = 1
m_2 = 2

t_1 = 0
t_2 = 0

for x in range(0, 3):
    xs.append(x) 
    ys_1.append((m_1 * x) + t_1)
    ys_2.append((m_2 * x) + t_2)
    
plt.plot(xs, ys_1, "r-") # Rote Gerade = g1
plt.text(1.375, 1.1, "g1")
plt.plot(xs, ys_2, "b-") # Blaue Gerade = g2
plt.text(0.9, 2.15, "g2")
plt.grid(True)
plt.show()
```

## y-Achsenabschnitt "t"

Die bisher behandelten Geraden waren Ursprungsgeraden. Diese gehen immer durch den Nullpunkt (0|0) und haben somit den Wert 0 für t

Die Variable t in der Formel $y = m * x + t$ beschreibt die Verschiebung der Gerade an der y-Achse. Wird ein Wert der größer als 0 für $t$ ist angegeben, verschiebt sich die Gerade nach oben. Ist der $t$ kleiner als 0 verschiebt sie sich nach unten.

## Beispiel

Folgender Code beschreibt zwei Geraden. Verändere die Werte der Variablen

t_1 und t_2

und vergleiche beide Geraden miteinander. Anhand der sich verändernden y-Werte bei x = 0 erkennst du, wie die Geraden im Koordinatensystem verschoben werden.

```python
import matplotlib.pyplot as plt

xs = []
ys_1 = []
ys_2 = []

m_1 = 1
m_2 = 1

t_1 = 2
t_2 = -1

for x in range(0, 3):
    xs.append(x) 
    ys_1.append((m_1 * x) + t_1)
    ys_2.append((m_2 * x) + t_2)
    
plt.plot(xs, ys_1, "r-") # Rote Gerade
plt.text(1.375, 3.15, "g1")
plt.plot(xs, ys_2, "b-") # Blaue Gerade
plt.text(1.375, 0.65, "g2")
plt.grid(True)
plt.show()
```
