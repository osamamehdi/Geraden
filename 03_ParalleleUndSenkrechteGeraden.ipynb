## Parallele Geraden

Ist die Steigung von zwei Geraden **gleich**, dann sind diese Parallel. 

Sie unterscheiden sich **nur** im y-Achsenabschnitt t. 

### Beispiel

Gegeben sind die Geraden $g: y = \frac{2}{3} - 1$ und $h: y = \frac{2}{3}x + 2$ 

```python
import matplotlib.pyplot as plt

xs = []
ys_g = []
ys_h = []

# Steigung der beiden Geraden
m = 2/3

# y-Achsenabschnitt t der Geraden g
t_g = -1
# y-Achsenabschnitt t der Geraden h
t_h = 2

for x in range(-2, 3):
    xs.append(x)
    ys_g.append((m * x) + t_g)
    ys_h.append((m * x) + t_h)
    
plt.plot(xs, ys_g, "r-")
plt.plot(xs, ys_h, "b-")
plt.grid(True)
plt.show()
```

## Senkrechte Geraden

Geraden die Senkrecht aufeinander stehen nennt man auch orthogonal. 

Wenn das Produkt der Steigungen **-1** ergibt, stehen zwei Geraden senkrecht aufeinander. Du kannst auch herausfinden ob zwei Geraden senkrecht aufeinander stehen indem du den Winkel zwischen den beiden Geraden überprüfst. Dieser sollte immer **90°** sein.

## Rechenbeispiel

### **Beispiel 1**
Gegeben: $g: y = -4x + 4$ und $h: y = \frac{1}{4}x + 1$

Gesucht: Sind diese Geraden senkrecht aufeinander?

Lösung:

$m_g = -4$; $m_h = \frac{1}{4}$

Wenn $m_g * m_h = -1$ dann stehen die Geraden senkrecht aufeinander.

$-4 * \frac{1}{4} = -1$ -> Geraden sind senkrecht!



### **Beispiel 2**
Gegeben: $g: y = 2x + 4$ und $h: y = 0.5x + 6$

Gesucht: Sind diese Geraden senkrecht aufeinander?

Lösung:

$m_g = 2$; $m_h = 0.5$

Wenn $m_g * m_h = -1$ dann stehen die Geraden senkrecht aufeinander.

$2 * 0.5 = 1$ -> Geraden sind **nicht** senkrecht!
