# Geradengleichungen

Allgemeine Formel (Normalform) $y = m * x + t$

Diese besteht aus insgesamt 3 Komponenten:
m, x und t. 
Wird der Wert einer dieser Komponenten verändert, hat das Auswirkungen darauf, wie die Gerade aussieht.

## Beispiel

Folgender Code beschreibt eine einfache Gerade. 
Die sogenannten Variablen 

**m_1, t_1 und x_s** 

beschreiben die oben genannten Komponenten der Geradengleichung.
Ändere die Variablen

**m_1** und **t_1**

ab und drücke oben auf "Run" und beobachte wie sich die Gerade auf unterschiedliche Werte für die Variablen **m_1** und **t_1** verändert.

```python
import matplotlib.pyplot as plt

xs = []
ys_1 = []

m_1 = 1
t_1 = 0

# Berechnung der y-Werte der darzustellenden Geraden
for x in range(0, 3):
    xs.append(x) # Wertebereich der x-Achse festlegen
    ys_1.append((m_1 * x) + t_1)
    
plt.plot(xs, ys_1, "r-") # Rote Gerade
plt.grid(True)
plt.show()
```

## Steigung

Die Variable **"m"** beschreibt die Steigung der Gerade. Diese lässt sich mit der Formel 

$m = \frac{x_2 - x_1}{y_2 - y_1}$ 

berechenen.

## Beispiel

Folgender Code beschreibt zwei Geraden. Verändere die Werte der Variablen 

**m_1** und **m_2**

und vergleiche beide Geraden miteinander. Anhand der unten eingeblendeten Geraden erkennst du, wie sich die Steigung auf die Geraden auswirkt.

```python
import matplotlib.pyplot as plt

xs = []
ys_1 = []
ys_2 = []

m_1 = 1
m_2 = 2

t_1 = 0
t_2 = 0

# Berechnung der y-Werte der darzustellenden Geraden
for x in range(0, 3):
    xs.append(x) # Wertebereich der x-Achse festlegen
    ys_1.append((m_1 * x) + t_1)
    ys_2.append((m_2 * x) + t_2)
    
plt.plot(xs, ys_1, "r-") # Rote Gerade
plt.plot(xs, ys_2, "b-") # Blaue Gerade
plt.grid(True)
plt.show()
```

## y-Achsenabschnitt "t"

Die bisher behandelten Geraden waren Ursprungsgeraden. Diese gehen immer durch den Nullpunkt (0|0) und haben somit den Wert 0 für t

Die Variable t in der Formel $y = m * x + t$ beschreibt die Verschiebung der Gerade an der y-Achse. Wird ein Wert der größer als 0 ist angegeben, verschiebt sich die Gerade nach oben. Ist der Wert kleiner als 0 verschiebt sie sich nach unten.

## Beispiel

Folgender Code beschreibt zwei Geraden. Verändere die Werte der Variablen

t_1 und t_2

und vergleiche beide Geraden miteinander. Anhand der sich verändernden y-Werte bei x = 0 erkennst du, wie die Geraden im Koordinatensystem verschoben werden.

```python
import matplotlib.pyplot as plt

xs = []
ys_1 = []
ys_2 = []

m_1 = 1
m_2 = 1

t_1 = 2
t_2 = -1

# Berechnung der y-Werte der darzustellenden Geraden
for x in range(0, 3):
    xs.append(x) # Wertebereich der x-Achse festlegen
    ys_1.append((m_1 * x) + t_1)
    ys_2.append((m_2 * x) + t_2)
    
plt.plot(xs, ys_1, "r-") # Rote Gerade
plt.plot(xs, ys_2, "b-") # Blaue Gerade
plt.grid(True)
plt.show()
```
